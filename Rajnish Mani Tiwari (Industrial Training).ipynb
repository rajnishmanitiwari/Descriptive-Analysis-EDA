```python
# this function will give an integer value to every Fuel Type 
def trans_fuel_type(x):
  if x == 'gas':
    return 0
  if x == 'diesel':
    return 1
```

```python
df['trans_Make'] = df['make'].apply(trans_make)   # assigning integer values to every company and add it in dataframe
df['trans_FuelType'] = df['fuel-type'].apply(trans_fuel_type)   # assiging integer values to every fuel type and add it in dataframe
df
```

```python
# taking input and output
# INPUT IS ALWAYS TWO DIMENSIONAL
x1 = df.iloc[:, 26:28].values
x1
```

```python
# OUTPUT IS ALWAYS ONE DIMENSIONAL
y1 = df.iloc[:,25].values
y1
```

```python
# train and test variable
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1,y1)
# random_state is very similer to random.seed(0)
# By default - 75% training data goes to training variables
# and rest 25% is used as the testing data

# The moment we run the train_test_split cell the data of x variable is divided into
# x_train(75%) and x_test(25%) similalry the data of variable y is divided into y_train(75%) and y_test(25%)
```

```python
print(x1.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y1.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# NORMALIZATION OR SCALING
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
# APPLY CLASSIFIER / REGRESSOR / CLUSTERER - CLASSIFIER
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# fitting the model
model.fit(x_train, y_train)   # we are using the x_train and y_train values to train/fit the model
```

```python
# Predictor variable
y_pred1 = model.predict(x_test)
y_pred1    # predicted output values
```

```python
y_test  # actual output values
```

```python
# to find out the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred1, y_test) * 100
```
